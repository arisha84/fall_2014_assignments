### First Model: Logistical regression

```python
(X_train, X_test,
 y_train, y_test) = cv.train_test_split(x, y,
                                        test_size=.2)
```

```python
logreg = lm.LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
y_train_p = logreg.predict(X_train)
```

```python
y_predicted = logreg.predict(X_test)
```

```python
t = y_predicted - y_test
ones = 0;

for i in t:
    if i == 1 or i == -1:
        ones+=1
        
print ones*1.0 / len(t)
df_t = pd.DataFrame(t, columns=['match'])
gb = df_t.groupby('match')
print gb.size()
```

```python
t = y_predicted + y_test
ones = 0;

for i in t:
    if i == 1:
        ones+=1
        
print ones*1.0 / len(t)
df_t = pd.DataFrame(t, columns=['match'])
gb = df_t.groupby('match')
print gb.size()
```
